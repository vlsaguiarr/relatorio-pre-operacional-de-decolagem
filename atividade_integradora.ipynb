##Algoritmo de verificação realizado em python.

Algoritmo identifica se possui algum parâmetro fora do ideal para realizar o lançamento, e retorna com "PRONTO PARA DECOLAR" ou "DECOLAGEM ABORTADA"

Neste algoritmo adicionei uma coluna para indicar qual condição seria responsável pela decisão "DECOLAGEM ABORTADA".

Adicionei também informações relevantes para melhoria do algoritmo e dos parâmetros.

A partir da análise assistida por IA, pude também aperfeiçoar esses parâmetros e meu dataset.

```python
from os import abort
dados=pd.read_csv('dados_ficticios.csv')

def decisão(row):
  cond_integridade = (row["integridade_estrutural"] == 1)
  cond_energia = (row["niveis_energia_percentual"] >= 70.0)
  cond_pressao = (row["pressao_tanques_psi"] >= 50.0) and (row["pressao_tanques_psi"] <= 180.0)
  cond_modulo_status = (row["status_modulos_criticos"] == str("Normal"))
  cond_temperatura_interna = (row["temperatura_interna_celsius"] >= 20.0) and (row["temperatura_interna_celsius"] <= 25.0)
  cond_temperatura_externa = (row["temperatura_externa_celsius"] >= -152.0) and (row["temperatura_externa_celsius"] <= 125.0)

  failed_conditions = []
  if not cond_integridade:
    failed_conditions.append('Integridade Estrutural')
  if not cond_energia:
    failed_conditions.append('Níveis de Energia')
  if not cond_pressao:
    failed_conditions.append('Pressão dos Tanques')
  if not cond_modulo_status:
    failed_conditions.append('Status dos Módulos Críticos')
  if not cond_temperatura_interna:
    failed_conditions.append('Temperatura Interna')
  if not cond_temperatura_externa:
    failed_conditions.append('Temperatura Externa')

  if not failed_conditions:
    return 'PRONTO PARA DECOLAR', ''
  else:
    return 'DECOLAGEM ABORTADA', ', '.join(failed_conditions)

dados[['decisao', 'condicao_falha']] = dados.apply(lambda row: pd.Series(decisão(row)), axis=1)

ready = dados[dados['decisao'] == 'PRONTO PARA DECOLAR'].shape[0]

abort = dados[dados['decisao'] == 'DECOLAGEM ABORTADA'].shape[0]

print(f"Quantidade de decisões 'PRONTO PARA DECOLAR': {ready}")
print(f"Quantidade de decisões 'DECOLAGEM ABORTADA': {abort}")

print(f"Porcentagem de decisões 'PRONTO PARA DECOLAR': {ready / dados.shape[0] * 100:.2f}%")
print(f"Porcentagem de decisões 'DECOLAGEM ABORTADA': {abort / dados.shape[0] * 100:.2f}%")

print(f"pProporção (1 para x) 'PRONTO PARA DECOLAR': {dados.shape[0] / ready:.2f}")
print(f"Proporção (1 para x) 'DECOLAGEM ABORTADA': {dados.shape[0] / abort:.2f}")

display(dados.head(16))
```

#Análise energética

Realizada a partir das informações de Capacidade total (kwh), Carga Atual (%), Consumo estimado na decolagem e perdas energéticas.

Com isso conseguimos dividir a Carga atual (%) por 100 e multiplicar pela Capacidade total (kwh) para extrairmos a informação de Carga atual (kwh)

A partir dela, subtraimos com as informações de Consumo estimado na decolagem e perdas energéticas para obtermos qual será a Autonomia energética inicial (kwh).

```python
import pandas as pd

autonomiaI = pd.read_csv('analise_energia.csv')

autonomiaI['Carga atual (kwh)'] = autonomiaI['Capacidade total (kwh)'] *  (autonomiaI['Carga atual (%)'] / 100)

autonomiaI['Autonomia inicial (kwh)'] = autonomiaI['Carga atual (kwh)'] - autonomiaI['Consumo estimado na decolagem (kwh)'] - autonomiaI['Perdas energéticas (kwh)']

display(autonomiaI.head(20))
```

```python
# @title
import pandas as pd
from IPython.display import display, Markdown, HTML
from collections import Counter

# --- Análise Geral da Decisão de Lançamento ---
display(Markdown("## Análise da Decisão de Lançamento"))
display(HTML(f"""
<div style="border: 1px solid #ccc; padding: 15px; border-radius: 5px; background-color: #f9f9f9;">
    <p><b>Total de Registros Avaliados:</b> {dados.shape[0]}</p>
    <p><b>Decisões 'PRONTO PARA DECOLAR':</b> {ready} ({ready / dados.shape[0] * 100:.2f}%)</p>
    <p><b>Decisões 'DECOLAGEM ABORTADA':</b> {abort} ({abort / dados.shape[0] * 100:.2f}%)</p>
    <p><b>Proporção 'PRONTO PARA DECOLAR':</b> 1 em cada {dados.shape[0] / ready:.2f} registros</p>
    <p><b>Proporção 'DECOLAGEM ABORTADA':</b> 1 em cada {dados.shape[0] / abort:.2f} registros</p>
</div>
"""))

# --- Detalhamento de Anomalias e Riscos para Decolagens Abortadas ---
display(Markdown("## Detalhamento de Anomalias e Riscos (Decolagens Abortadas)"))

aborted_data = dados[dados['decisao'] == 'DECOLAGEM ABORTADA']

if not aborted_data.empty:
    all_failed_conditions = []
    for conditions_str in aborted_data['condicao_falha']:
        if conditions_str: # Ignora strings vazias
            all_failed_conditions.extend([cond.strip() for cond in conditions_str.split(',')])

    if all_failed_conditions:
        condition_counts = Counter(all_failed_conditions)
        sorted_conditions = condition_counts.most_common()

        display(Markdown("### Frequência das Condições de Falha:"))
        counts_str = "<ul style=\"list-style-type: disc; margin-left: 20px;\">"
        for condition, count in sorted_conditions:
            counts_str += f"<li><b>{condition}</b>: {count} ocorrências</li>"
        counts_str += "</ul>"
        display(HTML(counts_str))

        display(Markdown("### Sugestões de Risco e Próximos Passos:"))
        risks_suggestions = []

        if 'Integridade Estrutural' in condition_counts:
            risks_suggestions.append(
                "<b>Integridade Estrutural:</b> A alta frequência de falhas na integridade estrutural indica a necessidade de revisões rigorosas nos processos de fabricação, inspeção e manutenção dos componentes estruturais da nave. Isso pode envolver o uso de materiais mais resistentes, design aprimorado ou tecnologias de detecção de falhas mais avançadas."
            )
        if 'Níveis de Energia' in condition_counts:
            risks_suggestions.append(
                "<b>Níveis de Energia:</b> Falhas nos níveis de energia sugerem que o sistema de gerenciamento de energia ou as fontes de energia (baterias, células de combustível) podem não estar adequados. É crucial otimizar a capacidade de armazenamento, monitoramento do consumo e garantir que as reservas de energia atendam aos requisitos mínimos para todas as fases da missão."
            )
        if 'Pressão dos Tanques' in condition_counts:
            risks_suggestions.append(
                "<b>Pressão dos Tanques:</b> Anomalias na pressão dos tanques indicam problemas potenciais com válvulas, selos, sensores ou até mesmo a composição dos propelentes. A falha nesse parâmetro é crítica e pode levar a eventos catastróficos. Uma revisão detalhada dos sistemas de propulsão e pressurização é fundamental."
            )
        if 'Status dos Módulos Críticos' in condition_counts:
            risks_suggestions.append(
                "<b>Status dos Módulos Críticos:</b> Se módulos críticos não estão em status 'Normal', isso aponta para falhas em subsistemas essenciais (navegação, comunicação, controle de ambiente, etc.). É vital implementar um diagnóstico mais profundo para identificar a raiz dos problemas nesses módulos e garantir sua redundância ou robustez."
            )
        if 'Temperatura Interna' in condition_counts:
            risks_suggestions.append(
                "<b>Temperatura Interna:</b> Temperaturas internas fora da faixa ideal podem indicar falhas nos sistemas de controle térmico, superaquecimento de componentes eletrônicos ou problemas de isolamento. Manter a temperatura dentro dos limites operacionais é crucial para o desempenho e a vida útil dos equipamentos a bordo."
            )
        if 'Temperatura Externa' in condition_counts:
            risks_suggestions.append(
                "<b>Temperatura Externa:</b> Variações extremas na temperatura externa, se não gerenciadas, podem causar estresse térmico em materiais e componentes. A adequação dos materiais e dos sistemas de isolamento para suportar o ambiente externo é uma área chave para investigação e aprimoramento."
            )

        if risks_suggestions:
            suggestions_str = "<ul style=\"list-style-type: circle; margin-left: 20px;\">"
            for suggestion in risks_suggestions:
                suggestions_str += f"<li>{suggestion}</li>"
            suggestions_str += "</ul>"
            display(HTML(suggestions_str))
        else:
            display(Markdown("Não foram encontradas condições de falha específicas para gerar sugestões detalhadas."))

    else:
        display(Markdown("Não foram encontradas falhas específicas para as decolagens abortadas que contenham condições detalhadas."))
elif abort > 0:
    display(Markdown("Houve decolagens abortadas, mas o campo 'condicao_falha' está vazio ou não detalhado."))
else:
    display(Markdown("Não houve decolagens abortadas neste conjunto de dados."))

# --- Amostra dos Dados com Decisão de Lançamento ---
display(Markdown("## Amostra dos Dados com Decisão de Lançamento"))
display(dados.head(10).style.set_table_styles([{'selector': '', 'props': [('border', '1px solid #eee')]}, {'selector': 'th', 'props': [('background-color', '#007bff'), ('color', 'white')]}]))
```
